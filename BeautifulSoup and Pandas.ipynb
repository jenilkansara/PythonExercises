## 1.3

```python
def get_data(i):
    row=pokedex_values_list[i]
    poke_name = row.find(class_ = 'ent-name').text
    poke_url = row.find( class_ = 'ent-name'),
    poke_url=poke_url[0]['href']
    poke_url = 'https://pokemondb.net' + poke_url
    
    poke_type = row.find_all(class_ = 'type-icon')
    poke_type_list =[]
    
    poke_points = row.find(class_ = 'cell-total').text
    
    poke_vals = row.find_all(class_ = 'cell-num')
    poke_vals_list=[]
    
    for element in poke_vals:
        poke_vals_list.append(element.text)
        
    for element in poke_type:
        poke_type_list.append(element.text)
    
    row_list=[poke_name, poke_url, poke_type_list, poke_points, poke_vals_list]
    row_df = pd.DataFrame(row_list).T
    return row_df
```

```python
df = pd.DataFrame()
for i in range(1,len(pokedex_values_list)):
    df=df.append(get_data(i))
```

```python
types = ['Grass','Poison','Fire','Flying','Dragon','Water','Bug','Normal','Dark','Electric', 'Psychic', 'Ground', 'Ice', 'Steel', 'Fairy', 'Fighting', 'Rock', 'Ghost']
```

```python
scores_df = pd.DataFrame(columns=['Attack', 'Defense'], index=types)
scores_df
```

```python
for i in types:
    scores_df['Attack'][i]=df[df[i]==1].Attack.mean()
    scores_df['Defense'][i]=df[df[i]==1].Defense.mean()
```

```python
scores_df
```

```python
scores_df.Attack=scores_df.Attack.astype(float)
scores_df.Defense=scores_df.Defense.astype(float)
```

```python
maxval_attack= 0
maxval_defense=0
minval_attack = 1000
minval_defense=1000
min_attack_type = ''
max_attack__type= ''
min_defense_type= ''
max_defense_type= ''

for index, row in scores_df.iterrows():
    if row['Attack']>maxval_attack:
        maxval_attack=row['Attack']
        max_attack__type=index
    if row['Attack']<minval_attack:
        minval_attack=row['Attack']
        min_attack__type=index
    if row['Defense']>maxval_defense:
        maxval_defense=row['Defense']
        max_defense_type=index
    if row['Defense']<minval_defense:
        minval_defense=row['Defense']
        min_defense_type=index
print('Mix Attack Avaerage is ' + str(minval_attack) + ' for ' + min_attack__type)
print('Max Attack Avaerage is ' + str(maxval_attack) + ' for ' + max_attack__type)
print('Min Defense Avaerage is ' + str(minval_defense) + ' for ' + min_defense_type)
print('Max Defense Avaerage is ' + str(maxval_defense) + ' for ' + max_defense_type)
```
